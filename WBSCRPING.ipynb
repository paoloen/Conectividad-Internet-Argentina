```python
AccesoVelBajadaIFijoLocal = pd.read_csv('123456.csv')  # Reemplaza 'ruta_del_archivo.csv' con la ruta de tu archivo CSV
```

```python
rango_columnas = AccesoVelBajadaIFijoLocal.columns[5:83]  # Columnas del 4 al 82

# Convertir el rango de columnas a números enteros
AccesoVelBajadaIFijoLocal[rango_columnas] = AccesoVelBajadaIFijoLocal[rango_columnas].replace(',', '', regex=True).astype(float)
```

```python
rango_columnas = AccesoVelBajadaIFijoLocal.columns[4:83]  # Columnas del 4 al 82

# Calcular la suma de las columnas y guardar el resultado en una nueva columna
AccesoVelBajadaIFijoLocal['Cant_Hogares'] = AccesoVelBajadaIFijoLocal[rango_columnas].sum(axis=1)
```

```python
AccesoVelBajadaIFijoLocal.info()
```

```python
columnas_seleccionadas = AccesoVelBajadaIFijoLocal.columns[4:83]  # Columnas del 4 al 82

# Cambiar el nombre de las columnas conservando solo la primera parte antes del primer espacio
nuevos_nombres = [col.split(" ")[0] for col in columnas_seleccionadas]
AccesoVelBajadaIFijoLocal.rename(columns=dict(zip(columnas_seleccionadas, nuevos_nombres)), inplace=True)
```

```python
nombres_columnas = AccesoVelBajadaIFijoLocal.columns

# Crear una nueva fila con los nombres de las columnas repetidos
nueva_fila = pd.DataFrame([nombres_columnas], columns=nombres_columnas)

# Concatenar la nueva fila con el dataset existente
AccesoVelBajadaIFijoLocal = pd.concat([nueva_fila, AccesoVelBajadaIFijoLocal], ignore_index=True)
```

```python
AccesoVelBajadaIFijoLocal = AccesoVelBajadaIFijoLocal.drop(index=1).reset_index(drop=True)
```

```python
columnas_seleccionadas = AccesoVelBajadaIFijoLocal.columns[6:83]  # Columnas del 6 al 82

# Multiplicar las filas por la primera fila y sumarlas en una nueva columna
AccesoVelBajadaIFijoLocal['SUMA'] = AccesoVelBajadaIFijoLocal[columnas_seleccionadas].mul(AccesoVelBajadaIFijoLocal[columnas_seleccionadas].iloc[0]).sum(axis=1)
```
